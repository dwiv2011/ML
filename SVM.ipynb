```python
import pandas as pd
import itertools
import os 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
pd.set_option('display.max_columns', None,"display.max_rows", None)
import matplotlib.pyplot as plt
import sklearn
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier   #1vs1 & 1vsRest Classifiers
from sklearn.svm import SVC
import string
```

```python
data=pd.read_csv("letter-recognition.csv")
data.head()
```

```python
train_df,test_df = train_test_split(data,test_size=0.3,stratify=data['letter'],random_state=11)
x_train=train_df.iloc[:,1:]
y_train=train_df.iloc[:,:1]
x_test=test_df.iloc[:,1:]
y_test=test_df.iloc[:,:1]

'''sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)'''
```

```python

uni=list(string.ascii_uppercase)
Ker=['linear','linear','linear','poly','poly','rbf']
c=[1,5,10,5,5,5]
d=[3,3,3,2,3,3]
acc=[]
ltr=[]
kern=[]
for i in range(6):
    model = SVC(kernel =Ker[i], C = c[i],degree=d[i])
    ovo = OneVsRestClassifier(model)
    ovo.fit(x_train, y_train.values.ravel())
    ovo_pred = ovo.predict(x_test)
    for s in uni:
        act_y=y_test["letter"]==s
        acc_=act_y.sum()
        pred_y=ovo_pred==s 
        cpred=(act_y==pred_y)&(act_y==True)
        pred_=cpred.sum()
        accur=pred_/acc_
        acc.append(accur)
        ltr.append("letter -"+s)
        if Ker[i]=="poly":
            kern.append(Ker[i]+"-"+str(c[i])+"-"+str(d[i]))
        else:
            kern.append(Ker[i]+"-"+str(c[i]))
        
        
        
        
    
    
```

```python
p=pd.DataFrame(zip(acc,ltr,kern))
p.columns=["Accuracy","Letter Picked","Kernel paramters"]
```

```python
df_final=p.pivot(index="Letter Picked", columns="Kernel paramters", values="Accuracy")
```

```python
def Accuracy(predictions, y):
    predictions = predictions.reshape(predictions.shape[0], 1)
    y = y.reshape(y.shape[0], 1)
    return (predictions == y).sum() / y.shape[0]
```

```python
# form all pairwise combinations
uni=list(string.ascii_uppercase)
comb = itertools.combinations(uni, 2)
accuracy,pos_class,Neg_class=[], [],[]
for (i,j) in comb:
    f=train_df[(train_df['letter']==i) | (train_df['letter']==j)]
    x_train=f.iloc[:,1:]
    y_train=f["letter"]
    clf = SVC(kernel='linear',C=5)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    acc=Accuracy(y_pred,np.array(y_test))
    accuracy.append(acc)
    pos_class.append(i)
    Neg_class.append(j)
    
    
```

```python
p=pd.DataFrame(zip(pos_class,Neg_class,accuracy))
p.columns=["Positive Class","Negative Class","Linear (C=5)"]
p
```
