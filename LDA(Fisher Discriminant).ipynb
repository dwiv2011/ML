```python
import pandas as pd
import itertools
import os 
import numpy as np
pd.set_option('display.max_columns', None,"display.max_rows", None)
```

```python
train=pd.read_csv(".//digit-recognizer//train.csv")
test=pd.read_csv(".//digit-recognizer//test.csv")
```

```python


pair1=[]
pair2=[]
stuff = list(range(10))
for L in range(0, len(stuff)+1):
    for subset in itertools.combinations(stuff, 2):
        pair1.append(subset[0])
        pair2.append(subset[1])
```

```python
#del data
for i in range(len(pair1)):
    
    data_A=train[train['label']==pair1[i]]
    data_B=train[train['label']==pair2[i]]
    mean_A=data_A.mean()
    mean_B=data_B.mean()
    diff=(mean_A-mean_B)**2
    varA=data_A.var()
    varB=data_B.var()
    countA=data_A.shape[0]
    countB=data_B.shape[0]
    fisher=diff/(countB*varB+countA*varA)
    fisher2=fisher.fillna(0)
    fisher2a=fisher[1:]
    m=fisher2a.mean()
    s=fisher2a.std()
    fisher3=fisher2a.apply(lambda x: (x-m)/s)
    fisher4=fisher3.fillna(0)
    st=str(pair1[i])+"-"+str(pair2[i])
    out1=pd.concat([pd.Series([st]), fisher4])
    if i==0:
        sample=fisher4
        data=pd.DataFrame(out1).transpose()
    else:
        data=data.append(out1,ignore_index=True)
```
